# Credit Card Dataset Overview

This notebook explores the credit card fraud dataset and performs basic exploratory data analysis.

```python
credit_df.isnull().sum()
```

```python
print("Duplicates:", credit_df.duplicated().sum())
```

```python
credit_df['Class'].value_counts()
```

```python
credit_df['Class'].value_counts(normalize=True) * 100
```

```python
sns.countplot(x='Class', data=credit_df)

plt.title("Class Distribution")

plt.show()
```

```python
credit_df['Amount'].describe()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(credit_df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")

plt.show()
```

## Findings

- The dataset is highly imbalanced.
- Fraudulent transactions represent a very small percentage of observations.
- Transaction amounts vary significantly across observations.
